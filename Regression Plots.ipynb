```python
# The normal imports
import numpy as np
from numpy.random import randn
import pandas as pd

# Import the stats library from numpy
from scipy import stats

# These are the plotting modules and libraries we'll use:
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

# Command so that plots appear in the iPython Notebook
%matplotlib inline
```

```python
tips= sns.load_dataset("tips")
```

```python
tips.head()
```

```python
sns.lmplot("total_bill","tip",tips)

#Line is linear regression for this specific scatter plot
```

```python
sns.lmplot("total_bill","tip",tips,order=1,
           scatter_kws={"marker":"o","color":"seagreen"},
           line_kws={"linewidth":1,"color":"blue"})

#use kws dictionary to pass arguments to seaborn
#Use order for using different orders other than linear
```

```python

sns.lmplot("total_bill","tip",tips,fit_reg=False)
```

```python

tips["tip_pect"]=100*(tips["tip"]/tips["total_bill"])
```

```python
#Jitter is periodic signal added to the data so that we can see it better in this case

sns.lmplot("size","tip_pect",tips,x_jitter=0.1)
```

```python
sns.lmplot("size","tip_pect",tips,x_estimator=np.mean)
```

```python
#Differentiate sex by using by specifying markers
sns.lmplot("total_bill","tip_pect",tips,hue="sex",markers=["x","o"])
```

```python
sns.lmplot("total_bill","tip_pect",tips,hue="day")
```

```python
#Check out local regression
sns.lmplot("total_bill","tip_pect",tips,lowess=True,line_kws={"color":"black"})
```

```python
sns.regplot("total_bill","tip_pect",tips)
```

```python
fig, (axis1,axis2) =plt.subplots(1,2)

sns.regplot("total_bill","tip_pect",tips,ax=axis1)
sns.violinplot(tips["size"],tips["tip_pect"],ax=axis2, inner="quartile")
```
